```python
import numpy as np

import matplotlib.pyplot as plt

import pandas as pd
```

# Verinin import edilmesi

```python

veriler=pd.read_csv("maaslar_yeni.csv")
veriler.head(20)
```

#Verilerin x ve y olarak ayrılması

Unvan'ların bir karşılığı var zaten.. Sözel  karşlığı kaldırmak akıllıca olacaktır.




```python
veriler_yeni=veriler.drop(['unvan',"Calisan ID"] , axis=1)

veriler_yeni
```

# Model ve Algoritma oluşturma

Verilerin features ve target olarak ayrılması

```python
x=veriler_yeni.iloc[:,0:3]

y=veriler_yeni.iloc[:,-1:]

y
```

MODEL OLUŞTURMA

 Multiplear Linear Regression

```python
from sklearn.linear_model import LinearRegression

Lr=LinearRegression() #hemen bir kapsül oluşturalım
Lr.fit(x.values,y.values)
```

```python
CeoTahmini=Lr.predict([[10,10,100]]) #Ceo olup 10 yıldır çalışıp 100 puan alan MAAŞ TAHMİNİ

print(CeoTahmini)


MudurTahmini=Lr.predict([[7,10,100]]) #Müdür olup 10 yıldır çalışıp 1000 puan alan MAAŞ TAHMİNİ

print(MudurTahmini)

Mudur2Tahmini=Lr.predict([[11,17,100]])

Mudur2Tahmini
```

MLS'in P value değeri 

```python
import statsmodels.api as sm
model=sm.OLS(Lr.predict(x.values),x.values)
model.fit().summary()
```

Polinomsal Regresyon

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures   #bunu kullanıcaz.


poly_reg=PolynomialFeatures(degree=2) #bir kapsül oluşturdum ve degree(Derecesini 2 yapalım buna istediğimi verebilirim ama benim problemime uygun olmalı)

x_poly=poly_reg.fit_transform(x.values) #x'in değerlerini polinomal dünyaya çevirme.

print(x_poly)

    #şimdi yine linearRegression kapsülünü kullanarak. y=a^0+bx+cx^2 ye göre y'yi öğren dicez.

lin_reg2=LinearRegression()
lin_reg2.fit(x.values,y.values)

print(lin_reg2.predict([[10,10,100]]))#Ceo olup 10 yıldır çalışıp 100 puan alan MAAŞ TAHMİNİ
print(lin_reg2.predict([[7,10,100]]))#Müdür olup 10 yıldır çalışıp 100 puan alan MAAŞ TAHMİNİ
```

Polinamsal Regresyonun P Value Değeri 

```python
import statsmodels.api as sm

model = sm.OLS(lin_reg2.predict(x.values),x.values)

model.fit().summary()
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

r_dt=DecisionTreeRegressor(random_state=0) #bir kapsül oluşturalım

r_dt.fit(x.values,y.values) #x'ten y'leri öğren diyoruz.

print(r_dt.predict([[10,10,100]])) #Ceo olup 10 yıldır çalışıp 100 puan alan MAAŞ TAHMİNİ by Decision Tree
print(r_dt.predict([[7,10,100]])) #Müdür olup 10 yıldır çalışıp 1000 puan alan MAAŞ TAHMİNİ
```

Decision Tree'in P value Değeri..

```python
import statsmodels.api as sm
model=sm.OLS(r_dt.predict(x.values),x.values) #Decision Tree'in x.values tahminleri al, x'le karşılaştır.
model.fit().summary()
```

Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

randomfor_reg= RandomForestRegressor(random_state=0,n_estimators=10) #hemen bir kapsül oluşturalım. Ve bu kapsülün içine iki tane parametre vericem.
#bunlardan biri DecisionTree'de de olduğu gibi random_state=0, diğeri ise kaç tane decision tree çizeleceği olan (n_estimators=10)
randomfor_reg.fit(x.values,y.values) #x.values'den y.valuesi öğren
```

```python

print(randomfor_reg.predict([[10,10,100]])) #random_forest'ın  10 yıllık CEONUN ve 100 puan almış statüdeki bir kişiye vereceği maaş tahmini
print(randomfor_reg.predict([[7,10,100]])) #random_forest'ın 10 yıllık MÜDÜRÜN ve 100 puan almış statüdeki bir kişiye vereceği maaş tahmini
```

Random Forest'in P-Value değeri

```python
import statsmodels.api as sm
model=sm.OLS(randomfor_reg.predict(x.values),x.values) #randomforest'in x.values tahminleri al, x'le karşılaştır.
model.fit().summary()
```

Support Vector Regression


```python

from sklearn.svm import SVR

svr_reg=SVR(kernel='rbf') #bu ikisi arasındaki ilişkiyi kurmasını istiyoruz.
svr_reg.fit(x.values,y.values)

print(svr_reg.predict([[10,10,100]])) #10 yıllık CEONUN ve 100 puan almış statüdeki bir kişiye vereceği maaş tahmini


print(svr_reg.predict([[7,10,100]])) #10 yıllık CEONUN ve 100 puan almış statüdeki bir kişiye vereceği maaş tahmini

```
