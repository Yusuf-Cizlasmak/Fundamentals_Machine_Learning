```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
veriler=pd.read_csv("odev_tenis.csv")

veriler.head(20)
```

```python
#ödev olduğunu bildiğimiz için.. İlk kim ak kim kara onu bilmemiz gerekiyor. Büyük ihtimalle Target kolonum PLAY olacak. Diğerleri de features olacak. Ondan sonra outlook'a da bir labelEncoder uygulucam.
# 
#  Ayrıca windy de aynı sistemi uygulucam ama dikkat DUMMY variables tuzağı olabilir.
```

## VERİLERİN ÖN İŞLEMESİ

```python
featuresDf=veriler.drop(columns=["play"],axis=1) #şimdi sadece features'larım kaldı..

#şimdi labelEncoder- Ve OneHotEncoding yöntemi..
outlook=veriler.iloc[:,0:1].values
outlook
```

```python
#şimdi de LabelEncoder ve OneHotEncoder yöntemi

from sklearn import preprocessing
le= preprocessing.LabelEncoder()

outlook[:,0]=le.fit_transform(veriler.iloc[:,0])

print(outlook)

ohe=preprocessing.OneHotEncoder()
outlook=ohe.fit_transform(outlook).toarray()

print(outlook)
```

```python
#Windy'nin OneHotEncoder yöntemi ..Trick noktaya dikkat sadece bir kolon alınacak.

windyDf=veriler.iloc[:,-2:-1].values


windyDf[:,-1]=le.fit_transform(veriler.iloc[:,-2:-1])


windyDf=ohe.fit_transform(windyDf).toarray()
print(windyDf)
```

```python
playDf=veriler.iloc[:,-1:].values


playDf[:,-1]=le.fit_transform(veriler.iloc[:,-1:])


playDf=ohe.fit_transform(playDf).toarray()
print(playDf)
```

```python

DigerFeatures=veriler.iloc[:,1:3]
DigerFeatures
```

```python
#Numpy dizilerinin DataFrame dönüşmesi..

sonuc=pd.DataFrame(data=outlook,index=range(14),columns=["overcast","rainy","sunny"])

sonuc1=pd.DataFrame(data=windyDf[:,1:2],index=range(14),columns=["windy"])

sonuc2=pd.DataFrame(data=playDf[:,1:2],index=range(14),columns=["play"])
```

```python
sonuc #Rüzgarlı günler 1 olarak belirlenirken, olmayan günler 0 olarak belirlenmiştir.
```

```python
sonuc2 #Oynanabilirler 1 oynanamazlar 0 olarak belirlenmiştir.
```

## Verileri birleştirme işlemi

```python
s=pd.concat([sonuc,DigerFeatures],axis=1)
s2=pd.concat([sonuc1,s],axis=1)
```

```python
s3=pd.concat([s2,sonuc2],axis=1)
```

```python
GeriS3=s2.drop(columns=["windy","sunny","temperature"],axis=1)
GeriS3
```

## VERİYİ TEST VE TRAIN OLARAK AYIRMA VE MODEL HAZIRLAMA
