```python
#eksik verileri tamamlamak bazen çıkarmak, bazen de tamamlamak gerekebilir.


from sklearn.impute import SimpleImputer

#sci-kit learn anlamına gelir ,sklearn = scikit learn

imputer=SimpleImputer(missing_values=np.nan,strategy="mean") #missing valueları np.nan olarak seçtik çünkü onları doldurucaz.

#peki strategy olarak da ortalamayı kullanarak doldursun.

```

```python
Yas=veriler.iloc[:,1:4].values #iloc=index location , ilk satırın başından itibaren 4.sütunu alır.

print(Yas) #görüldüğü üzere bu bir dizi.
```

```python
imputer=imputer.fit(Yas[:,1:4])  #fit öğrenilmesi için kullanılan fonksiyondur.


#öğrendikten sonra da transform fonksiyonu ile nan değerlerini doldurmasını isticez.

Yas[:,1:4]=imputer.transform(Yas[:,1:4])



print(Yas)
```

```python
#Kategorik veri dönüşümleri 


ulke=veriler.iloc[:,0:1].values #ilk parametrem tüm satırları alırken, virgülden sonrası sütunu alır.
#ayrıca DataFrame'den diziye dönüştürdük.
print(ulke)
```

```python
#Label encoding : kategorik verileri sayısal verilere çevirir.

#kütüphanemizi import edelim.

from sklearn import preprocessing

labeLEnconing=preprocessing.LabelEncoder()


ulke[:,0]=labeLEnconing.fit_transform(ulke[:,0])  #fit_transform ile öğrenme yapılırken transform ile değerleri dönüştürülür.

print(ulke)
```

```python
#BİR BAŞKA METHOD İSE oneHotEncoder


ohc=preprocessing.OneHotEncoder() 
ulke=ohc.fit_transform(ulke).toarray()

print(ulke)
```

```python


sonuc= pd.DataFrame(data=ulke,index= range(22),columns=["fr","tr","us"])

sonuc
```
