Fit yöntemi iki şey döndürür: sözde birliktelik kuralı indeksleri ve sözde birliktelik kuralı destekler.Daha önce açıkladığım gibi, çok fazla metrik olmayacak. Buradaki tek ilginç şey, aşağıdaki kodu kullanarak kural desteğine bakmaktır:
