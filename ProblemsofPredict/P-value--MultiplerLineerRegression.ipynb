```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Verilerin çekilmesi

veriler=pd.read_csv("veriler.csv")
#veriler.describe()


veriler.head(20)
```

## VERİNİN ÖN İŞLEMESİ

Dummy variables and trap konusunun kodları ve önemine binaen

```python
Yas=veriler.iloc[:,1:4].values #iloc=index location , ilk satırın başından itibaren 4.sütunu alır.

print(Yas) #görüldüğü üzere bu bir dizi.
```

```python
#Ülke ve cinsiyet kolonlarını regresyon problemi olduğu için sayısal değerlere çevirmem gerekiyor.
#bunun için LabelEncoder yönteminden yararlanıcam

#İlk önce gerçek veriden bir ayıralım.
ulke=veriler.iloc[:,0:1].values
ulke

#şimdi de LabelEncoder ve OneHotEncoder yöntemi

from sklearn import preprocessing
le= preprocessing.LabelEncoder()

ulke[:,0]=le.fit_transform(veriler.iloc[:,0])

print(ulke)
```

```python
ohe=preprocessing.OneHotEncoder()
ulke=ohe.fit_transform(ulke).toarray()

print(ulke)
```

```python
#şimdi aynısını cinsiyetler için yapalım..

c=veriler.iloc[:,-1:].values


c[:,-1]=le.fit_transform(veriler.iloc[:,-1:])


c=ohe.fit_transform(c).toarray()
print(c)
```

```python
#Numpy dizilerinin DataFrame dönüşmesi..

sonuc=pd.DataFrame(data=ulke,index=range(22),columns=["fr","tr","us"])

sonuc1=pd.DataFrame(data=Yas,index=range(22),columns=['boy','kilo','yas'])

sonuc2=pd.DataFrame(data=c[:,:1],index=range(22),columns=["cinsiyet"])


```

```python
#Birleştirmeden önce dikkat etmemiz gereken husus "Dummy variables and Trap" konusu..


sonuc2  #almamız gerekiyor ki dummy variables tuzağına düşmemek adına.(ERKEKLERİ BAZ ALDIK.)
```

## VERİLERİN BİRLEŞTİRİLMESİ

```python
s=pd.concat([sonuc,sonuc1],axis=1)
s2=pd.concat([s,sonuc2],axis=1)


print(s2)
```

## VERİYİ TEST VE TRAIN OLARAK AYIRMA VE MODEL HAZIRLAMA


Test ve Train olarak ayrılması

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(s,sonuc2,test_size=0.30,random_state=42)

```
